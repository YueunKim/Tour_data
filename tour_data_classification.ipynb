```python
# 키워드별로 개수를 구함 -> 대표 키워드 선정 
```

```python
keywords = []

for i in df.splited:
    for j in range(0,len(i)):
        a = i[j]
        keywords.append(a)
        
keywords
```

```python
# keywords의 모든 요소 개수 세기

dic = {}

for list_name in keywords:
    dic[list_name] = 0

for list_count in keywords:
    dic[list_count] = dic[list_count] + 1

print(dic)
```

```python
# 내림차순 정렬
dic_items = dic.items()
sort = sorted(dic_items, key=lambda x: x[1], reverse=True)
sort
```
